```python
import requests
import csv
import codecs
import json
import time



headers = {'User-Agent': 'api-test-agent'}
total_res = []
for page_num in range(0,20):
      url = "https://api.hh.ru/vacancies"
      params = {
        'specialization': "1",
        "text": "NAME:баз данных",
        'area': 113, # Поиск ощуществляется по вакансиям россии
        'page': page_num, # Индекс страницы поиска на HH
        'per_page': 100 # Кол-во вакансий на 1 странице
      }
      resp = requests.get(url, params=params,headers=headers).json()
      time.sleep(4)
      print(resp)
      if resp.get("items"):
         items=resp["items"]
         for item in items:
            vak_id=item["id"]
            vak_name=item["name"]
            vak_siti=item["area"]["name"] 
            salary_f=item["salary"]["from"] if item.get("salary") else ""#проверка на наличие salary(если в первом элеиенте нет, то пусто)
            salary_to=item["salary"]["to"] if item.get("salary") else ""
            salary_currency=item["salary"]["currency"] if item.get("salary") else ""
            data=item["published_at"]
            data2 = data[:10].replace("-", ".").split(".")[::-1]
            data3= ".".join(data2) 
            if vak_id: 
              spec_id = []
              key_list=[]
              url1=f'https://api.hh.ru/vacancies/{vak_id}'
          # print(url1)
              ones = requests.get(url1).json() 
              time.sleep(1) 
              key_list.append(','.join([one["name"] for one in ones ["key_skills"] if one.get("name")]))
              spec_id.append(' '.join([one["name"] for one in ones ["specializations"] if one.get("name")]))
           #print(ones)             
              result={
                "id":vak_id,
                "spec":spec_id,
                "name":vak_name,
                "skills":key_list,
                "city":vak_siti,
                "zp_ot":salary_f,
                "zp_do":salary_to,
                "val":salary_currency,
                "published_at":data3,
                 }              
        #print(data)

              total_res.append(result) 
            # print(result)       
with open('database_all.csv', 'w') as csvfile:
       fieldnames = total_res[0].keys()
       writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
       writer.writeheader() 
       for spec in total_res:
                writer.writerow(spec) #при записи строк ссылка на результирующий список 
            #сначала цикл по его элементам (спискам results), затем по словарям каждого списка
```
